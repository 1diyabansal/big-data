```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from sklearn.linear_model import LinearRegression
```

```python
x=[2,4,6,8]
y=[3,7,5,10]
x=np.array([2,4,6,8]).reshape(-1,1)
y=np.array([3,7,5,10]).reshape(-1,1)


lr=LinearRegression()
lr.fit(x,y)

x_new=12
x_new = np.array(x_new).reshape(-1, 1)
y_pred=lr.predict(x_new)
print(f"Predicted value of y for x = {x_new}: {y_pred}")
```

```python

import matplotlib.pyplot as plt
# Create a scatter plot of the training data
plt.scatter(x, y, color='blue', label='Training data')

# Plot the regression line
plt.plot(x, lr.predict(x), color='red', linewidth=3, label='Regression line')

# Plot the prediction for the new data point
plt.scatter(x_new, y_pred, color='green', marker='x', label='Prediction')

# Add labels and title to the plot
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```
